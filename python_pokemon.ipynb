```python
# 讓除以0不要error的兩個方法

def safe_div(a, b):
    # a>0 & b>=0
    if not b == 0:
        return a/b
    else:
        return float("inf")
    
def safe_div2(a, b, eps=1e-10):
    # a>0 & b>=0
    return a/(b+eps)
```

```python
# 輕鬆虐菜
# 輸入 s = "#006"(str), 回傳 s 打人戰鬥倍率大於1的列表
def pkFreeWin(s):
    
    # 建立poke字典
    f = open("pokemon.txt", "r")
    poke_dic = {}
    poke_name_dic = {}
    poke_list = []
    for line in f:
        line_split = line.strip().split("\t")
        poke_dic[line_split[0]] = line_split[3:]
        poke_name_dic[line_split[0]] = line_split[1]
        poke_list += [line_split[0]]
    f.close()

    # 建立屬性互打字典
    f = open("attribute.txt", "r")
    attr_dic = {}
    attr_list = []
    for line in f:
        line_split = line.strip().split("\t")
        attr_list += [line_split[0]]
        attr_dic[line_split[0]] = line_split[1:]     
    f.close()
    
    # 任意屬性打屬性
    def att2att(a, b, attr_dic, attr_list):
        return float(attr_dic[a][attr_list.index(b)])
    
    #任意poke打poke
    def pk2pk(attack, enemy, poke_dic, attr_dic, attr_list):

        attack = poke_dic[attack]
        enemy = poke_dic[enemy]
        #print("attack_attr{} -> enemy_attr{}".format(attack, enemy))

        attack_rates =[]
        for i in attack:
            attack_rate = 1.0
            for j in enemy:
                attack_rate *= att2att(i, j, attr_dic, attr_list)
            attack_rates.append(attack_rate)
        best_attack = max(attack_rates)

        enemy_rates = []
        for i in enemy:
            enemy_rate = 1.0
            for j in attack:
                enemy_rate *= att2att(i, j, attr_dic, attr_list)
            enemy_rates.append(enemy_rate)
        best_enemy = max(enemy_rates)

        if best_enemy == 0:
            return float("inf")
        else:
            return float(best_attack / best_enemy)
    
    your_ans = []
    for i in poke_list:
        point = pk2pk(s, i, poke_dic, attr_dic, attr_list)
        if point > 1.0:
            your_ans.append(poke_name_dic[i])
    
    return your_ans

enemy = '#006'

pkFreeWin(enemy)
```

```python
# 全員6v6
# team1 = [6隻], team2 = [6隻]
# default 派出 index= 0 的出來

def pk6vs6(team1, team2):
    
    # 建立poke字典(key = 中文, value = 屬性)
    f = open("pokemon.txt", "r")
    poke_dic = {}
    for line in f:
        line_split = line.strip().split("\t")
        poke_dic[line_split[1]] = line_split[3:]
    f.close()

    # 建立屬性互打字典
    f = open("attribute.txt", "r")
    attr_dic = {}
    attr_list = []
    for line in f:
        line_split = line.strip().split("\t")
        attr_list += [line_split[0]]
        attr_dic[line_split[0]] = line_split[1:]     
    f.close()
    
    # 任意屬性打屬性(a, b = "屬性")
    def att2att(a, b, attr_dic, attr_list):
        return float(attr_dic[a][attr_list.index(b)])
    
    #任意poke打poke(attack, enemy = "怪獸名")
    def pk2pk(attack, enemy, poke_dic, attr_dic, attr_list):

        attack = poke_dic[attack]
        enemy = poke_dic[enemy]
        #print("attack_attr{} -> enemy_attr{}".format(attack, enemy))

        attack_rates =[]
        for i in attack:
            attack_rate = 1.0
            for j in enemy:
                attack_rate *= att2att(i, j, attr_dic, attr_list)
            attack_rates.append(attack_rate)
        best_attack = max(attack_rates)

        enemy_rates = []
        for i in enemy:
            enemy_rate = 1.0
            for j in attack:
                enemy_rate *= att2att(i, j, attr_dic, attr_list)
            enemy_rates.append(enemy_rate)
        best_enemy = max(enemy_rates)

        if best_enemy == 0:
            return float("inf")
        else:
            return float(best_attack / best_enemy)
    
    def find_best_member(team, enemy, poke_dic, attr_dic, attr_list):
        try:
            team_rates = []
            for member in team:
                team_rates.append(pk2pk(member, enemy, poke_dic, attr_dic, attr_list))
            #print("team_rates: {}".format(team_rates))
            max_idx = team_rates.index(max(team_rates))
            return team[max_idx]
        except:
            return
    
    while (len(team1) != 0) and (len(team2) != 0):
        t1_point = pk2pk(team1[0], team2[0], poke_dic, attr_dic, attr_list)
        t2_point = pk2pk(team2[0], team1[0], poke_dic, attr_dic, attr_list)
        if t1_point > t2_point:
            team2.pop(0)
            t2_best = find_best_member(team2, team1[0], poke_dic, attr_dic, attr_list)
            if t2_best in team2:
                team2.remove(t2_best)
                team2.insert(0, t2_best)
        elif t2_point > t1_point:
            team1.pop(0)
            t1_best = find_best_member(team1, team2[0], poke_dic, attr_dic, attr_list)
            if t1_best in team1:
                team1.remove(t1_best)
                team1.insert(0, t1_best)
        elif t1_point == t2_point:
            team1.pop(0)
            team2.pop(0)
                
    if (len(team1) == 0) and (len(team2) == 0):
        return "Deuce"
    elif (len(team1) == 0) and (len(team2) != 0):
        return team2
    elif (len(team2) == 0) and (len(team1) != 0):
        return team1
    
team1 = ["妙蛙花", "噴火龍", "水箭龜", "急凍鳥", "閃電鳥", "火焰鳥"]
team2 = ["小火龍", "傑尼龜", "妙蛙種子", "多邊獸", "超夢", "皮卡丘"]

team1 = ["小火龍", "傑尼龜", "妙蛙種子", "小火龍", "傑尼龜", "妙蛙種子"]
team2 = ["小火龍", "傑尼龜", "妙蛙種子", "小火龍", "傑尼龜", "妙蛙種子"]


pk6vs6(team1, team2)
```
